```python

# Define the path to the folder where your filtered GeoJSON files are located
filtered_folder = 'filtered_geojson'

# Dictionary to store the GeoDataFrames with their corresponding filenames as keys
gdf_dict = {}

# Loop through each file in the filtered folder
for filename in os.listdir(filtered_folder):
    if filename.endswith('.geojson'):
        # Construct the full file path
        filepath = os.path.join(filtered_folder, filename)
        
        # Read the GeoJSON file into a GeoDataFrame
        gdf = gpd.read_file(filepath)
        
        # Store the GeoDataFrame in the dictionary using the filename (without extension) as the key
        key = os.path.splitext(filename)[0]
        gdf_dict[key] = gdf
        
        print(f"Loaded {key}")

# Access individual GeoDataFrames
print(gdf_dict['pipelines_ca'].head())
print(gdf_dict['basins_ru'].head())
```

### Now ill give these numbers context with global averages by country `



```python
# Step 1: Load the GeoJSON files
def load_geojson_files(folder):
    gdf_dict = {}
    for filename in os.listdir(folder):
        if filename.endswith('.geojson'):
            filepath = os.path.join(folder, filename)
            gdf_name = os.path.splitext(filename)[0]
            gdf = gpd.read_file(filepath)
            gdf_dict[gdf_name] = gdf
            print(f"Loaded {gdf_name}")
    return gdf_dict
```

```python
# Step 2: Calculate average counts grouped by MD_Country
def calculate_average_counts(gdf_dict):
    average_counts = {}
    for gdf_name, gdf in gdf_dict.items():
        if 'MD_Country' in gdf.columns:
            country_counts = gdf.groupby('MD_Country').size()
            average_count = country_counts.mean()
            average_counts[gdf_name] = average_count
            print(f"Average count of features per MD_Country in {gdf_name}: {average_count}")
    return average_counts
```

```python
# Load the GeoJSON files
gdf_dict = load_geojson_files('ogdata')

# Calculate and print the average counts
average_counts = calculate_average_counts(gdf_dict)

# Print the results
print("\nAverage count of features per MD_Country for all GeoDataFrames:")
for gdf_name, avg_count in average_counts.items():
    print(f"{gdf_name}: {avg_count}")
```
